# Unificación de datasets ambiental y general del año 2020

```python
# importar librerias
import pandas as pd
print("Carga de librerías exitosa")
```

```python
# Carga datasets ambiental y general
df_ambiental_2020 = pd.read_csv("/content/drive/MyDrive/Bootcamp Análisis de Datos/Grupo Q/data/procesados/ambiental/2020_eas_ambiental.csv")
df_ambiental_2020.head()
```

```python
df_general_2020 = pd.read_csv("/content/drive/MyDrive/Bootcamp Análisis de Datos/Grupo Q/data/procesados/general/2020_eas_general.csv")
df_general_2020.head()
```

Se unifican los datasets mediante un inner join con el id de la empresa

```python
# Merge de los dos datasets por campo id_empresa
df_servicios_unificado_2020 = pd.merge(df_ambiental_2020, df_general_2020, on="id_empresa", how="inner")
df_servicios_unificado_2020.head()
```

```python
# inspecciona el dataframe unificado
df_servicios_unificado_2020.info()
```

Existen columnas con el mismo nombre en ambos datasets, por lo tanto pandas les modifica el nombre, a continuación se deja una sola con el nombre original.

```python
# Elimina columna periodo repetida
df_servicios_unificado_2020.drop(columns=["periodo_y"], inplace=True)
df_servicios_unificado_2020.rename(columns={"periodo_x":"periodo"}, inplace=True)
print(df_servicios_unificado_2020.columns)
```

## Verificación de consistencia de id_empresa e id_sector
Se comparan los id_sector e id_sector_general, y si son inconsistentes se busca la razón para realizar ajustes o eliminar registros.

```python
# Extrae id_sector diferentes
condicion_diferente_Sector = df_servicios_unificado_2020['id_sector'] != df_servicios_unificado_2020['id_sector_general']
df_servicios_unificado_2020.loc[condicion_diferente_Sector,
                                ['id_empresa',
                                 'id_sector',
                                 'total_ingresos',
                                 'id_sector_general']]
```

Se evidencia un patrón en las inconsistencias presentadas, el id_sector LN, aparece con un caracter de más "LN1", se ajusta para que coincidan

```python
# Corrige id_sector_general LN1 a LN
df_servicios_unificado_2020['id_sector_general'] = df_servicios_unificado_2020['id_sector_general'].mask(df_servicios_unificado_2020['id_sector_general']=='LN1', 'LN')
condicion_diferente_Sector = df_servicios_unificado_2020['id_sector'] != df_servicios_unificado_2020['id_sector_general']
df_servicios_unificado_2020.loc[condicion_diferente_Sector,
                                ['id_empresa',
                                 'id_sector',
                                 'total_ingresos',
                                 'id_sector_general']]
```

```python
# Elimina registros con incongruencias en id_sector e id_sector_general
df_servicios_unificado_2020 = df_servicios_unificado_2020[df_servicios_unificado_2020['id_sector'] == df_servicios_unificado_2020['id_sector_general']]
df_servicios_unificado_2020.shape
```

Se eliminaron satisfactoriamente los registros con id_sector inconsistentes
