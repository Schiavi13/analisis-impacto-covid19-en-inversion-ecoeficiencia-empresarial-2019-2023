# Unificación de datasets ambiental y general del año 2019

```python
# importar librerias
import pandas as pd
print("Carga de librerías exitosa")
```

```python
# Carga datasets ambiental y general
df_ambiental_2019 = pd.read_csv("/content/drive/MyDrive/Bootcamp Análisis de Datos/Grupo Q/data/procesados/ambiental/2019_eas_ambiental.csv")
df_ambiental_2019.head()
```

```python
df_general_2019 = pd.read_csv("/content/drive/MyDrive/Bootcamp Análisis de Datos/Grupo Q/data/procesados/general/2019_eas_general.csv")
df_general_2019.head()
```

Se unifican los datasets mediante un inner join con el id de la empresa

```python
# Merge de los dos datasets por campo id_empresa
df_servicios_unificado_2019 = pd.merge(df_ambiental_2019, df_general_2019, on="id_empresa", how="inner")
df_servicios_unificado_2019.head()
```

```python
# inspecciona el dataframe unificado
df_servicios_unificado_2019.info()
```

## Modificación de tipos de datos

Se ajustan los tipos de datos de las columnas

```python
df_servicios_unificado_2019 = df_servicios_unificado_2019.astype(
    {
        'gastos_totales': 'float64',
        'gasto_gestion_amb': 'float64',
        'total_ingresos': 'float64',
        'personal_remunerado': 'int64',
        'gasto_remuneracion_personal': 'float64',
        'gasto_personal_total': 'float64',
        'personal_mujeres': 'int64',
        'personal_hombres': 'int64',
        'personal_ocupado': 'int64',
        'imp_ind_com': 'float64',
        'imp_renta_equidad': 'float64',
        'imp_otros': 'float64',
        'gasto_servicios_publicos': 'float64',
        'gasto_energia_electrica': 'float64',
        'gasto_gas_natural': 'float64'
    }
)
df_servicios_unificado_2019.info()
```

Existen columnas con el mismo nombre en ambos datasets, por lo tanto pandas les modifica el nombre, a continuación se deja una sola con el nombre original.

```python
# Elimina columna periodo repetida
df_servicios_unificado_2019.drop(columns=["periodo_y"], inplace=True)
df_servicios_unificado_2019.rename(columns={"periodo_x":"periodo"}, inplace=True)
print(df_servicios_unificado_2019.columns)
```

## Verificación de consistencia de id_empresa e id_sector
Se comparan los id_sector e id_sector_general, y si son inconsistentes se busca la razón para realizar ajustes o eliminar registros.

```python
# Extrae id_sector diferentes
condicion_diferente_Sector = df_servicios_unificado_2019['id_sector'] != df_servicios_unificado_2019['id_sector_general']
df_servicios_unificado_2019.loc[condicion_diferente_Sector,
                                ['id_empresa',
                                 'id_sector',
                                 'total_ingresos',
                                 'id_sector_general']]
```

Se evidencia que el sector "LN1" en id_sector_general tiene un número extra, se puede ajustar a "LN" para que coincida sin modificar los datos. Por otro lado, existen id_sector que no tienen relación, se aislan y se eliminan del análisis estos registros.

```python
# Corrige id_sector_general LN1 a LN
df_servicios_unificado_2019['id_sector_general'] = df_servicios_unificado_2019['id_sector_general'].mask(df_servicios_unificado_2019['id_sector_general']=='LN1', 'LN')
condicion_diferente_Sector = df_servicios_unificado_2019['id_sector'] != df_servicios_unificado_2019['id_sector_general']
df_servicios_unificado_2019.loc[condicion_diferente_Sector,
                                ['id_empresa',
                                 'id_sector',
                                 'total_ingresos',
                                 'id_sector_general']]
```

```python
# Elimina registros con incongruencias en id_sector e id_sector_general
df_servicios_unificado_2019 = df_servicios_unificado_2019[df_servicios_unificado_2019['id_sector'] == df_servicios_unificado_2019['id_sector_general']]
df_servicios_unificado_2019.shape
```

```python
# Verifica id_sector
condicion_diferente_Sector = df_servicios_unificado_2019['id_sector'] != df_servicios_unificado_2019['id_sector_general']
df_servicios_unificado_2019.loc[condicion_diferente_Sector,
                                ['id_empresa',
                                 'id_sector',
                                 'total_ingresos',
                                 'id_sector_general']]
```

Se eliminaron satisfactoriamente los registros con id_sector inconsistentes
