# Unificación de datasets ambiental y general del año 2022

```python
# importar librerias
import pandas as pd
print("Carga de librerías exitosa")
```

```python
# Carga datasets ambiental y general
df_ambiental_2022 = pd.read_csv("/content/drive/MyDrive/Bootcamp Análisis de Datos/Grupo Q/data/procesados/ambiental/2022_eas_ambiental.csv")
df_ambiental_2022.head()
```

```python
df_general_2022 = pd.read_csv("/content/drive/MyDrive/Bootcamp Análisis de Datos/Grupo Q/data/procesados/general/2022_eas_general.csv")
df_general_2022.head()
```

Se unifican los datasets mediante un inner join con el id de la empresa

```python
# Merge de los dos datasets por campo id_empresa
df_servicios_unificado_2022 = pd.merge(df_ambiental_2022, df_general_2022, on="id_empresa", how="inner")
df_servicios_unificado_2022.head()
```

```python
# inspecciona el dataframe unificado
df_servicios_unificado_2022.info()
```

Existen columnas con el mismo nombre en ambos datasets, por lo tanto pandas les modifica el nombre, a continuación se deja una sola con el nombre original.

```python
# Elimina columna periodo repetida
df_servicios_unificado_2022.drop(columns=["periodo_y"], inplace=True)
df_servicios_unificado_2022.rename(columns={"periodo_x":"periodo"}, inplace=True)
print(df_servicios_unificado_2022.columns)
```

## Verificación de consistencia de id_empresa e id_sector
Se comparan los id_sector e id_sector_general, y si son inconsistentes se busca la razón para realizar ajustes o eliminar registros.

```python
# Extrae id_sector diferentes
condicion_diferente_Sector = df_servicios_unificado_2022['id_sector'] != df_servicios_unificado_2022['id_sector_general']
df_servicios_unificado_2022.loc[condicion_diferente_Sector,
                                ['id_empresa',
                                 'id_sector',
                                 'total_ingresos',
                                 'id_sector_general']]
```

Se evidencia que el sector "LN1" en id_sector_general tiene un número extra, se puede ajustar a "LN" para que coincida sin modificar los datos. Por otro lado, existen id_sector que no tienen relación, se aislan y se eliminan del análisis estos registros.

```python
# Corrige id_sector_general LN1 a LN
df_servicios_unificado_2022['id_sector_general'] = df_servicios_unificado_2022['id_sector_general'].mask(df_servicios_unificado_2022['id_sector_general']=='LN1', 'LN')
condicion_diferente_Sector = df_servicios_unificado_2022['id_sector'] != df_servicios_unificado_2022['id_sector_general']
df_servicios_unificado_2022.loc[condicion_diferente_Sector,
                                ['id_empresa',
                                 'id_sector',
                                 'total_ingresos',
                                 'id_sector_general']]
```

## Manejo de rechazados
Extrae los registros inconsistentes a un csv aparte.

```python
# Crea nuevo dataframe con datos inconsistentes
condicion_diferente_Sector = df_servicios_unificado_2022['id_sector'] != df_servicios_unificado_2022['id_sector_general']
df_rechazados_unificado_2022 = df_servicios_unificado_2022.loc[condicion_diferente_Sector,
                                ['id_empresa',
                                 'id_sector',
                                 'sector_desc',
                                 'id_sector_general']]
df_rechazados_unificado_2022
```

```python
# Exporta el dataframe con datos inconsistentes a csv
df_rechazados_unificado_2022.to_csv("/content/drive/MyDrive/Bootcamp Análisis de Datos/Grupo Q/data/procesados/unificados/rechazados/2022_rechazados_eas_unificado.csv", index=False)
```
